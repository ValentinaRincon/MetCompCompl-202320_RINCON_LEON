```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def f1(x,y):
    return (x**2)-(y**2)+2*x
```

```python
funcion = f1(1.,1.)
```

```python
a = [1.,1.]
```

```python
Dx = lambda f,x,y,h=1e-5: (f(x+h,y) - f(x-h,y))/(2*h)
Dy = lambda f,x,y,h=1e-5: (f(x,y+h) - f(x,y-h))/(2*h)
    
```

```python
Gradient = lambda f,x,y: np.array([Dx(f,x,y),Dy(f,x,y)])
gradiente = Gradient(f1,a[0],a[1])
```

```python
x = sym.Symbol ('x',real=True)
y = sym.Symbol ('y',real=True)
```

```python
V = np.array([[x-1.],[y-1.]])
```

```python
V
```

```python
z = (funcion + (np.dot(gradiente,V)))
```

```python
z
```

```python
def plano (x,y):
    
    return (4*x**2) - 2*y
```

```python
x1 = np.linspace(-5, 5, 100)
y1 = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x1, y1)

Z = f1(X, Y)
tangente = plano(X,Y)
```

```python
figura = plt.figure()

ax = figura.add_subplot(111, projection = '3d')

ax.plot_surface(X,Y,Z, cmap = 'cool',alpha = 0.5)
ax.plot_surface(X,Y,tangente, cmap = 'jet',alpha = 0.5)


ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

plt.show()
```

```python
figura = plt.figure()
ax = figura.add_subplot(111, projection='3d')

# Trazar las superficies con etiquetas
surf1 = ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.5, label='Función')
surf2 = ax.plot_surface(X, Y, tangente, cmap='jet', alpha=0.5, label='Plano')

# Etiquetas de ejes
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")

# Obtener las líneas de las leyendas y los nombres de las etiquetas
lineas = [surf1, surf2]
nombres = [l.get_label() for l in lineas]

# Añadir la leyenda a la figura
figura.legend(lineas, nombres)

plt.show()
```

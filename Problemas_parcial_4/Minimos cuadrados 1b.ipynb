```python
import matplotlib.pyplot as plt
import numpy as np
```

Realice una b√∫squeda iterativa entre -5 <= x <= 5 y  -5 <= y <= 5 con un paso de h = 0.01 para encontrar la menor distancia posible del problema. Grafique la distancia.

```python
X = np.arange(-5,5,0.01)
Y = np.arange(-5,5,0.01)
```

```python
x = np.array([2,1,1])
y = np.array([-1,2,1])

b = np.array([2,1,4])
```

```python
matriz = np.zeros((3,2))
matriz[:,0] = x
matriz[:,1] = y
```

```python
matriz_distancias = np.zeros((len(X),len(Y)))

for i in range(len(X)):
    
    for j in range(len(Y)):
        
        r = np.array([X[i],Y[j]])
        
        distancia = np.linalg.norm(np.dot(matriz,r)-b)
        
        matriz_distancias[i][j] = distancia
```

Ahora, se crea una cuadricula para ver la figura graficada.

```python
X,Y = np.meshgrid(X,Y)
```

```python
figura = plt.figure()
ax1 = figura.add_subplot(111,projection="3d")
ax1.plot_surface(X,Y,matriz_distancias,cmap = "coolwarm")
ax1.set_xlabel("eje x")
ax1.set_ylabel("eje y")
ax1.set_zlabel("Distancia")
```

```python
print("La distancia m√≠nima usando este m√©todo es: {}".format(np.min(matriz_distancias)))
```

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

Calcular la probabilidad que n personas (n <= 365) tengan fechas distintas de cumpleaños, escribir la fórmula general del cálculo. Grafique la probabilidad P(n <= 80) como función de n.

Para comenzar, hay que tener en cuenta que para realizar este cálculo la idea es garantizar que en cada día cumpla una sola persona. Para ello, hay que dividir constantemente el número de casos favorables sobre los casos posibles. En este caso, el número de personas que no tienen una fecha de cumpleaños asignada sobre el número de días, 365 días. Cabe resaltar que hay al asignar a cada persona una fecha de cumpleaños, el número de casos favorables disminuye una unidad.

```python
p = 1

casos_posibles = 365

for i in range(casos_posibles):
    
    p_p = (casos_posibles-i)/casos_posibles
    
    p *= p_p
    
porcentaje_probabilidad = p*100
```

```python
print("La probabilidad que existe de cumplir en fechas distintas años en un grupo de 365 personas es: " +str(p) + ". El procentaje de esta probabilidad es: " + str(porcentaje_probabilidad))
```

```python
p_80personas = []

personas = np.linspace(0,80)

p_80 = 1

for i in range(len(personas)):
    
    if i <= 80:
        
        p_80p = (casos_posibles-i)/casos_posibles
        
        p_80 *= p_80p
        
        p_80personas.append(p_80)
```

```python
plt.scatter(personas,p_80personas, color = "pink")
plt.title("Probabilidad de que 80 personas cumplan en días distintos.")
plt.xlabel("Probabilidad de cumpleaños diferentes")
plt.ylabel("Probabilidad total")
plt.show()
```
